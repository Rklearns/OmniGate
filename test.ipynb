```python
import os
import json
import re

# Update this path if you are running the script from a different folder
file_path = "preprocessing/processed_multicancer/GS-BRCA/CNV_features.json"

print(f"Checking file: {file_path}")
if not os.path.exists(file_path):
    print("File not found! Please check your working directory.")
else:
    # 1. Print the raw first 5 lines to see the exact hidden characters
    print("\n--- RAW FIRST 5 LINES ---")
    with open(file_path, "r", encoding="utf-8") as f:
        raw_lines = [next(f, None) for _ in range(5)]
        for i, line in enumerate(raw_lines):
            if line is not None:
                print(f"Line {i}: {repr(line)}")

    # 2. Extract and count
    print("\n--- ATTEMPTING EXTRACTION ---")
    extracted_genes = []
    
    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read().strip()
        
        # Try JSON first
        try:
            data = json.loads(content)
            if isinstance(data, dict):
                # Sort keys as integers so "10" doesn't come before "2"
                for k in sorted(data.keys(), key=lambda x: int(x) if str(x).isdigit() else str(x)):
                    extracted_genes.append(data[k])
            elif isinstance(data, list):
                extracted_genes = data
            print("Status: Successfully parsed using JSON loader!")
            
        except json.JSONDecodeError:
            print("Status: Not standard JSON. Falling back to Regex text parsing...")
            # Matches anything inside double quotes or single quotes
            for line in content.split('\n'):
                match = re.search(r'["\']([^"\']+)["\']', line)
                if match:
                    extracted_genes.append(match.group(1))

    print(f"\nTotal genes extracted: {len(extracted_genes)}")
    print(f"First 10 genes: {extracted_genes[:10]}")
    
    # 3. Check against the .npy file dimension
    npy_path = "preprocessing/processed_multicancer/GS-BRCA/CNV_processed.npy"
    if os.path.exists(npy_path):
        import numpy as np
        data_shape = np.load(npy_path).shape
        print(f"\nActual CNV columns in .npy file: {data_shape[1]}")
        if data_shape[1] != len(extracted_genes):
            print("⚠️ MISMATCH! The number of features in the JSON does not match the columns in the NPY file.")
```
