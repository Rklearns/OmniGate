```python
import os
import numpy as np
import torch
import torch.nn as nn
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score
from sklearn.utils.class_weight import compute_class_weight

# =========================
# CONFIG
# =========================
SEED = 42
MAX_EPOCHS = 350
MIN_EPOCHS = 100
PATIENCE = 50
LR = 1e-3
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

torch.manual_seed(SEED)
np.random.seed(SEED)

BASE_DIR = "./processed_multicancer"
CANCER = "GS-BRCA"
DATA_DIR = os.path.join(BASE_DIR, CANCER)

# =========================
# LOAD DATA
# =========================
X_mrna  = np.load(os.path.join(DATA_DIR, "mRNA_processed.npy"))
X_mirna = np.load(os.path.join(DATA_DIR, "miRNA_processed.npy"))
X_cnv   = np.load(os.path.join(DATA_DIR, "CNV_processed.npy"))
X_methy = np.load(os.path.join(DATA_DIR, "Methy_processed.npy"))
y       = np.load(os.path.join(DATA_DIR, "labels.npy"))

num_classes = len(np.unique(y))

omics_data = {
    "mrna": X_mrna,
    "mirna": X_mirna,
    "cnv": X_cnv,
    "methy": X_methy
}

# =========================
# FOCAL LOSS
# =========================
class FocalLoss(nn.Module):
    def __init__(self, alpha, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, logits, targets):
        ce = nn.functional.cross_entropy(
            logits, targets, weight=self.alpha, reduction="none"
        )
        pt = torch.exp(-ce)
        loss = ((1 - pt) ** self.gamma) * ce
        return loss.mean()

# =========================
# MODEL
# =========================
class OmicsEncoder(nn.Module):
    def __init__(self, in_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, 512),
            nn.ReLU(),
            nn.BatchNorm1d(512),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU()
        )

    def forward(self, x):
        return self.net(x)


class MultiOmicsClassifier(nn.Module):
    def __init__(self, in_dims, num_classes):
        super().__init__()
        self.encoders = nn.ModuleDict({
            k: OmicsEncoder(v) for k, v in in_dims.items()
        })

        fusion_dim = 128 * len(in_dims)

        self.classifier = nn.Sequential(
            nn.Linear(fusion_dim, 256),
            nn.ReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(0.4),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(128, num_classes)
        )

    def forward(self, x_dict):
        z = []
        for k in self.encoders:
            z.append(self.encoders[k](x_dict[k]))
        z = torch.cat(z, dim=1)
        return self.classifier(z)

# =========================
# TRAIN ONE SPLIT
# =========================
def train_on_split(omics_data, y, train_idx, test_idx):
    X_train, X_test = {}, {}
    for k, X in omics_data.items():
        scaler = StandardScaler()
        X_train[k] = scaler.fit_transform(X[train_idx])
        X_test[k]  = scaler.transform(X[test_idx])

    y_train = y[train_idx]
    y_test  = y[test_idx]

    X_train_t = {k: torch.tensor(v, dtype=torch.float32).to(DEVICE)
                 for k, v in X_train.items()}
    X_test_t  = {k: torch.tensor(v, dtype=torch.float32).to(DEVICE)
                 for k, v in X_test.items()}

    y_train_t = torch.tensor(y_train, dtype=torch.long).to(DEVICE)
    y_test_t  = torch.tensor(y_test,  dtype=torch.long).to(DEVICE)

    classes = np.unique(y_train)
    alpha = compute_class_weight(
        "balanced", classes=classes, y=y_train
    )
    alpha = torch.tensor(alpha, dtype=torch.float32).to(DEVICE)

    criterion = FocalLoss(alpha=alpha, gamma=2.0)

    in_dims = {k: v.shape[1] for k, v in X_train.items()}
    model = MultiOmicsClassifier(in_dims, num_classes).to(DEVICE)

    optimizer = torch.optim.AdamW(
        model.parameters(), lr=LR, weight_decay=3e-4
    )

    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode="max", factor=0.5, patience=12, verbose=False
    )

    best_prec = 0.0
    best_state = None
    counter = 0

    for epoch in range(MAX_EPOCHS):
        model.train()
        optimizer.zero_grad()

        logits = model(X_train_t)
        loss = criterion(logits, y_train_t)
        loss.backward()
        optimizer.step()

        model.eval()
        with torch.no_grad():
            preds = model(X_test_t).argmax(dim=1).cpu().numpy()
            prec = precision_score(
                y_test, preds, average="weighted", zero_division=0
            )

        scheduler.step(prec)

        if prec > best_prec:
            best_prec = prec
            best_state = {k: v.cpu() for k, v in model.state_dict().items()}
            counter = 0
        else:
            counter += 1

        if epoch >= MIN_EPOCHS and counter >= PATIENCE:
            break

    model.load_state_dict({k: v.to(DEVICE) for k, v in best_state.items()})
    return best_prec

# =========================
# 5-FOLD CV
# =========================
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

precs = []
for i, (train_idx, test_idx) in enumerate(skf.split(X_mrna, y), 1):
    p = train_on_split(omics_data, y, train_idx, test_idx)
    precs.append(p)
    print(f"Fold {i}: PREC = {p:.4f}")

print(
    f"\nFINAL GS-BRCA PRECISION: "
    f"{np.mean(precs):.4f} ± {np.std(precs):.4f}"
)
```
